```python
import requests

articles = []

#1è‡³5é 
for i in range(1,5):
    #é ç¢¼è¨­å®šåƒæ•¸.format(i)
    res = requests.get('https://m.weibo.cn/api/container/getIndex?containerid=100103type%3D1%26q%3D%2523%25E6%25AD%25A6%25E6%25BC%25A2%25E8%2582%25BA%25E7%2582%258E&page_type=searchall&page={}'.format(i))
    jd = res.json()
    articles.extend([rec['mblog'] for rec in jd['data']['cards'] if rec.get('mblog')])
df = pandas.DataFrame(articles)
```
