#  python爬取微博武漢肺炎資料

```python
import requests

#先至微博取得Request URL,這裡僅爬取一頁
res = requests.get('https://m.weibo.cn/api/container/getIndex?containerid=100103type%3D1%26q%3D%2523%25E6%25AD%25A6%25E6%25BC%25A2%25E8%2582%25BA%25E7%2582%258E&page_type=searchall&page=3')
jd = res.json()
```

```python
import pandas

#整理成結構化資訊
df = pandas.DataFrame(jd['data']['cards'])
```

```python
articles = [rec['mblog'] for rec in jd['data']['cards'] if rec.get('mblog')]
df = pandas.DataFrame(articles)
```

```python
import requests

articles = []

#1至5頁
for i in range(1,5):
    #頁碼設定參數.format(i)
    res = requests.get('https://m.weibo.cn/api/container/getIndex?containerid=100103type%3D1%26q%3D%2523%25E6%25AD%25A6%25E6%25BC%25A2%25E8%2582%25BA%25E7%2582%258E&page_type=searchall&page={}'.format(i))
    jd = res.json()
    articles.extend([rec['mblog'] for rec in jd['data']['cards'] if rec.get('mblog')])
df = pandas.DataFrame(articles)
```

```python
#查看抓取幾頁資訊
len(df)
```

```python
df['text']
```

```python
from bs4 import BeautifulSoup

#將html的tag等的雜訊處理
def parseArticle(e):
    soup = BeautifulSoup(e, 'lxml')
    return soup.text

df['text'].map(parseArticle)
```
