# yahoo商城爬蟲 

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://tw.search.mall.yahoo.com/search/mall/product?kw=%E6%89%8B%E9%8C%B6&p=%E6%89%8B%E9%8C%B6&cid=hp&clv=0")
soup = BeautifulSoup(res.text)
   
```

針對特定class爬取
再將其存入dic中

```python
ary = []
for item in soup.select('.BaseGridItem__grid___2wuJ7'):
    #print(item.select('.BaseGridItem__grid___2wuJ7 em')[0].text,item.select('.BaseGridItem__title___2HWui')[0].text.strip(),item.select('.StoreGridItem__storeName___2dutX')[0].text)
    price = item.select_one('.BaseGridItem__grid___2wuJ7 em').text
    title = item.select_one('.BaseGridItem__title___2HWui').text.strip()
    store = item.select_one('.StoreGridItem__storeName___2dutX').text
    ary.append({'price':price,'title':title,'store':store})
```

```python
import pandas as pd
df = pd.DataFrame(ary)
```
