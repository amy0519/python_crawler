```python
import pandas as pd

df = pd.read_html('https://officeguide.cc/internet-country-code-table/')
df
```

```python
df1 = df[0]
df1
```

```python
df2 = df[1]
df2
```

```python
df3 = df[2]
df3
```

```python
frame = [df1,df2,df3]
df4=pd.concat(frame,axis=0,join='outer').fillna(0)
df4 = df4.reset_index(drop='true')

df4 = df4.rename(columns = {'中文':'國家(中)','英文':'國家(英)','三碼代碼':'國家代碼(三)','兩碼代碼':'國家代碼(二)'})
df4
```

```python
df4.to_excel('國家代碼.xlsx', sheet_name='1', index=False)
```
